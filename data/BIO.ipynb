```python
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
```

```python
# файл со всеми отзывами
# делим на id и тексты

texts, ids = [], []
with open('train_reviews.txt') as f:
    for line in f:
        text_id, text = line.rstrip('\r\n').split('\t')
        texts.append(text)
        ids.append(text_id)
```

```python
# делим на 2 набора: train и test

train_texts, test_texts, train_ids, test_ids = train_test_split(texts, ids)
```

```python
# файл с аспектами (выделенными сущностями)
# делим на 2 набора: train и test

train_aspects, test_aspects = [], []
with open('train_aspects.txt') as f:
    for line in f:
        line = line.rstrip('\r\n')
        text_id = line.split('\t')[0]
        if text_id in train_ids:
            train_aspects.append(line)
        if text_id in test_ids:
            test_aspects.append(line)
```

```python
# файл с категориями (оценки по каждой категории для каждого отзыва)
# делим на 2 набора: train и test

train_sentiment, test_sentiment = [], []
with open('train_cats.txt') as f:
    for line in f:
        line = line.rstrip('\r\n')
        text_id = line.split('\t')[0]
        if text_id in train_ids:
            train_sentiment.append(line)
        if text_id in train_ids:
            train_sentiment.append(line)
```

```python
train_ids, test_ids = [], []

with open('train_split_reviews.txt') as f:
    for line in f:
        text_id, text = line.rstrip('\r\n').split('\t')
        train_ids.append(text_id)

with open('test_reviews.txt') as f:
    for line in f:
        text_id, text = line.rstrip('\r\n').split('\t')
        test_ids.append(text_id)

print(len(train_ids), len(test_ids))
```

```python
test_sentiment = []

with open('train_cats.txt') as f:
    for line in f:
        line = line.rstrip('\r\n')
        text_id = line.split('\t')[0]
        if text_id in test_ids:
            test_sentiment.append(line)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import pymorphy2

import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
morph = pymorphy2.MorphAnalyzer()
stops = stopwords.words('russian')
```

```python
# токенизация + приведение к нижнему регистру + проверка на только буквы в составе +
# + лемматизация + удаление стоп-слов
def preprocessing(text):
    lemmas = []
    for word in word_tokenize(text):
      if word.isalpha():
        word = morph.parse(word.lower())[0]
        lemma = word.normal_form
        if lemma not in stops:
          lemmas.append(lemma)
    return ' '.join(lemmas)
```

```python
# создаем словарь, в котором ключ - id, значение - текст

reviews_ = {}
with open('train_split_reviews.txt') as f:
  for line in f:
    line = line.rstrip('\r\n').split('\t')
    reviews_[line[0]] = line[1]
```

```python
from collections import defaultdict
aspects_ = defaultdict(list)
```

```python
# словарь, в котором ключ - id, значение - список словарей (отдельный словарь для каждого аспекта)

with open('train_split_aspects.txt', 'r', encoding='utf-8') as f:
  for line in f:
    line = line.rstrip('\r\n').split('\t')
    keys = ('category', 'word', 'start', 'end', 'sentiment')
    aspects_[line[0]].append(dict(zip(keys, line[1:])))
```

```python
final_df_ = pd.DataFrame(columns=['id', 'bio_token', 'bio_tag'])
```

```python
# функция для bio-разметки

def bio(reviews, aspects, final_df):

  all_bio_token = []
  all_bio_tag = []

  for text_id, text in reviews.items():
    r_df = pd.DataFrame(columns=['id', 'bio_token', 'bio_tag'])
    bio_token = []
    bio_tag = []
    processed = nlp(text)
    for token in processed.iter_tokens():
      add = False
      for mention_dict in aspects[text_id]:
        mention_split = mention_dict['word'].split()
        if len(mention_split) == 1:
          if token.start_char == int(mention_dict['start']) and token.end_char == int(mention_dict['end']):
            bio_token.append(token.text)
            bio_tag.append('B-'+mention_dict['category'])
            add = True
            continue
        elif token.start_char == int(mention_dict['start']):
            bio_token.append(token.text)
            bio_tag.append('B-'+mention_dict['category'])
            add = True
            continue
        elif token.end_char == int(mention_dict['end']) or (token.start_char > int(mention_dict['start']) and token.end_char < int(mention_dict['end'])):
            bio_token.append(token.text)
            bio_tag.append('I-'+mention_dict['category'])
            add = True
            continue
      if not add:
        bio_token.append(token.text)
        bio_tag.append('O')
    r_df['bio_token'] = bio_token
    r_df['bio_tag'] = bio_tag
    r_df['id'] = text_id
    final_df = pd.concat([final_df, r_df])
    continue
  return final_df
```

```python
bio_df = bio(reviews_, aspects_, final_df_)

bio_df.to_csv('BIO_id_token_tag.csv', sep='\t', index=False)
```
